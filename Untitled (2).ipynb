### Moving Average

```python
df1=df[["petal width (cm)","sepal width (cm)"]]
df1.columns=["petal_width","sepal_width"]
```

```python
c1=np.array([random.uniform(min(df1["petal_width"]),max(df1["petal_width"])),
          random.uniform(min(df1["sepal_width"]),max(df1["sepal_width"]))])

c2=np.array([random.uniform(min(df1["petal_width"]),max(df1["petal_width"])),
          random.uniform(min(df1["sepal_width"]),max(df1["sepal_width"]))])
```

```python
plt.scatter(df1.petal_width,df1.sepal_width)
plt.scatter(c1[0],c1[1])
plt.scatter(c2[0],c2[1])
```

```python
l1=[]
l2=[]
for i in range(0,len(df1["petal_width"])):
    d1 = distance.euclidean(c1, df1.iloc[i,].values)
    d2 = distance.euclidean(c2, df1.iloc[i,].values)
    
    if d1<d2:
        l1.append(df1.iloc[i,].values)
    else:
        l2.append(df1.iloc[i,].values)

l1=pd.DataFrame(l1)
l2=pd.DataFrame(l2)
```

```python
plt.scatter(c1[0],c1[1])
plt.scatter(c2[0],c2[1])
plt.scatter(l1[0],l1[1])
plt.scatter(l2[0],l2[1])
```

### Iteration 2

```python
c1=np.array([np.average(l1[0]),np.average(l1[1])])

c2=np.array([np.average(l2[0]),np.average(l2[1])])
```

```python
max_itr=5
itr=0
while(itr <= max_itr):
    itr +=1
    
    ########
    c1=np.array([np.average(l1[0]),np.average(l1[1])])
    c2=np.array([np.average(l2[0]),np.average(l2[1])])    
    #########
    l1=[]
    l2=[]
    for i in range(0,len(df1["petal_width"])):
        d1 = distance.euclidean(c1, df1.iloc[i,].values)
        d2 = distance.euclidean(c2, df1.iloc[i,].values)

        if d1<d2:
            l1.append(df1.iloc[i,].values)
        else:
            l2.append(df1.iloc[i,].values)
    
    ########
    l1=pd.DataFrame(l1)
    l2=pd.DataFrame(l2)
    
    ########
    plt.scatter(c1[0],c1[1])
    plt.scatter(c2[0],c2[1])
    plt.scatter(l1[0],l1[1])
    plt.scatter(l2[0],l2[1])
    plt.show()
    time.sleep(10)
```
